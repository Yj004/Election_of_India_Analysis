### Documentation

##### Functions

```python
def convert_ratio(x):
  if x not in ['inf', '-inf', 'nan', 'Nan']:
    try:
      percentage = round(float(x) * 100, 2) # this would apply to 0 too, but would return 0 itself
      return percentage
    except ValueError:
      return x
  else:
    return x
```

```python
def replace_symbol(x):
  x = str(x)
  if '%' in str(x):
    x = x.replace('%', '').replace(' ', '')
    return x
  else:
    return x
```

```python
def not_numeric(x):
  if x not in ['inf', '-inf', 'NaN', 'nan']:
    try:
      float(x)
      return False
    except:
      return True
  else:
    return True
```

```python
def replacecomma(x):
  if type(x) == str:
    x = x.replace(',', '').replace('RU', '0')
    return x
  else:
    return x
```
